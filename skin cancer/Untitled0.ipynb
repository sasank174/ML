```python
import os
import shutil
import time


def func(path,target):
	for x in os.listdir(path):
		p = path + "/" + x
		t = target + "/" + x
		if os.path.isdir(p):
			print("============================"+ x +"============================")
			if not os.path.exists(t):
				os.makedirs(t)
			start_time = time.time()
			func(p,t)
			print("--- %s seconds ---" % (time.time() - start_time))
		elif os.path.isfile(p):
			shutil.copyfile(p, t)
```
